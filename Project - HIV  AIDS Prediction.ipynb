# HIV Dataset Analysis and Prediction

Country: The country associated with the data entry. Each row in your dataset likely      corresponds to a specific country.

Year: The year associated with the data entry. It indicates when the observation or measurement was recorded.

Count: Estimated number of people (all ages) living with HIV.

Count_median:The median value of the counts. Median is a   measure of central tendency that represents the middle value when the data is sorted.

Count_min: The minimum value of the counts, It indicates the lowest recorded count.

Count_max: The maximum value of the counts,It indicates the highest recorded count.

WHO Region:The World Health Organization (WHO) region associated with the country. WHO regions categorize countries based on geographic proximity or shared characteristics.

###  Here, in this data we take Country,Year,Count_median,Count_min,Count_max as indepetend features and Count as depended  or the target feature.

#### Based on the  dataset, target feature Count represents the count of events or incidents ,a numerical quantity, then regression models would be more appropriate. we can use regression analysis to predict the count of events based on other independent variables.

```python
#Slicing the dependent variable

x=h.drop("Count",axis=1).values
x
```

```python
#slicing the indpendent variables
y=h.iloc[:,2].values
y
```

```python
#spliting used in machine learning to evaluate the performance of a model on unseen data.

from sklearn.model_selection import train_test_split
```

```python
#To split a dataset into training and testing sets for model training and evaluation

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=10,random_state=10)
```

```python
# training the independent features

x_train
```

```python
#training the target variable
y_train
```

```python
#testing the features
x_test
```

```python
#testing the target variable
y_test
```

## To check the Accuracy using Linear Reg

```python
#importing Linear reg model from scikit learn
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()# fitting the model
regressor.fit(x_train,y_train)
```

```python
#the predictions made by the model on the input features
xpred = regressor.predict(x_train)
xpred
```

```python
ypred = regressor.predict(x_test)
ypred
```

```python
score= regressor.score(x_train,y_train)
score
```

```python
testingscore1= regressor.score(x_test,y_test)
testingscore1
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
print('mae',metrics.mean_absolute_error(y_test,ypred))  #Lower values of MAE indicate better model performance.
print('mse',metrics.mean_squared_error(y_test,ypred)) #Lower values of MSE indicate better model performance.
print('rmse',np.sqrt(metrics.mean_squared_error(y_test,ypred))) #Lower values of RMSE indicate better model performance.
print('rmae',np.sqrt(metrics.mean_absolute_error(y_test,ypred)))
r2s = r2_score(y_test,ypred)
print("r2s",r2s)  #Higher values of R2 indicate better model performance.
```
